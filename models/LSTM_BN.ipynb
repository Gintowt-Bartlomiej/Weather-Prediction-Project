```python
data_path = "../data/filtered_weather_data.csv"
data = pd.read_csv(data_path, parse_dates=['Formatted Date'], index_col='Formatted Date')

data['Hour'] = data.index.hour
data['Month'] = data.index.month

data.head(-5)
```

```python
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler
from sklearn.compose import ColumnTransformer
import numpy as np
```

```python
num_features = ['Temperature (C)', 'Apparent Temperature (C)', 'Humidity', 'Wind Speed (km/h)', 'Wind Bearing (degrees)', 'Pressure (millibars)', 'Visibility (km)', 'Hour', 'Month']
cat_features = ['Summary', 'Precip Type']
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', MinMaxScaler(), ['Temperature (C)', 'Apparent Temperature (C)', 'Humidity', 'Wind Speed (km/h)', 'Wind Bearing (degrees)', 'Pressure (millibars)', 'Visibility (km)', 'Hour', 'Month']),
        ('cat', OneHotEncoder(handle_unknown='ignore'), ['Summary', 'Precip Type'])
    ])
```

```python
test_data_processed = preprocessor.transform(test_data)
test_data_processed_df = pd.DataFrame(test_data_processed, columns=all_features)
test_data_processed_df
```

## Prepare Training Data

```python
def create_sliding_window(data, window_size, stride=1):
    X, Y = [], []
    for i in range(0, len(data) - window_size, stride):
        X.append(data.iloc[i:i + window_size, :].values)
        Y.append(data.iloc[i + window_size, 0])
    return np.array(X), np.array(Y)
```

```python
WINDOW_SIZE = 4
STRIDE = 1
```

```python
X_train, y_train = create_sliding_window(train_data_processed_df, WINDOW_SIZE, STRIDE)
X_test, y_test = create_sliding_window(test_data_processed_df, WINDOW_SIZE, STRIDE)
```

# Model

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
class BayesianLSTM(nn.Module):
    def __init__(self, n_features, output_length, batch_size, hidden_dim1=128, hidden_dim2=32, stacked_layers=2):
        super(BayesianLSTM, self).__init__()

        self.batch_size = batch_size

        self.hidden_size_1 = hidden_dim1
        self.hidden_size_2 = hidden_dim2
        self.stacked_layers = stacked_layers
        self.dropout_prob = 0.5

        self.lstm1 = nn.LSTM(
            n_features,
            self.hidden_size_1,
            num_layers=self.stacked_layers,
            batch_first=True
        )

        self.lstm2 = nn.LSTM(
            self.hidden_size_1,
            self.hidden_size_2,
            num_layers=self.stacked_layers,
            batch_first=True
        )

        self.fc = nn.Linear(self.hidden_size_2, output_length)
        self.loss_fn = nn.MSELoss()

        self.device = torch.device('cuda' if torch.cuda.is_available() else  torch.device("mps") if torch.backends.mps.is_available() else 'cpu')

        self.set_device(self.device)

    def forward(self, x):
        batch_size, seq_len, _ = x.size()

        hidden = self.init_hidden1(batch_size)
        output, _ = self.lstm1(x, hidden)
        output = F.dropout(output, p=self.dropout_prob, training=True)
        state = self.init_hidden2(batch_size)
        output, state = self.lstm2(output, state)
        output = F.dropout(output, p=self.dropout_prob, training=True)
        output = output[:, -1, :]
        y_pred = self.fc(output)
        return y_pred


    def init_hidden1(self, batch_size):
        hidden_state = torch.zeros(self.stacked_layers, batch_size, self.hidden_size_1, device=self.device, requires_grad=True)
        cell_state = torch.zeros(self.stacked_layers, batch_size, self.hidden_size_1, device=self.device, requires_grad=True)
        return hidden_state, cell_state

    def init_hidden2(self, batch_size):
        hidden_state = torch.zeros(self.stacked_layers, batch_size, self.hidden_size_2, device=self.device, requires_grad=True)
        cell_state = torch.zeros(self.stacked_layers, batch_size, self.hidden_size_2, device=self.device, requires_grad=True)
        return hidden_state, cell_state


    def loss(self, pred, truth):
        return self.loss_fn(pred, truth)

    def predict(self, X):
        return self(torch.tensor(X)).view(-1).detach().numpy()

    def set_device(self, device):
        self.to(device)
```

```python
from torch.utils.data import Dataset, DataLoader

class TimeSeriesDataset(Dataset):
    def __init__(self, X, Y):
        self.X = torch.tensor(X, dtype=torch.float32)
        self.Y = torch.tensor(Y, dtype=torch.float32)

    def __len__(self):
        return len(self.X)

    def __getitem__(self, idx):
        return self.X[idx], self.Y[idx]
```

```python
BATCH_SIZE = 32
EPOCHS = 25
```

```python
train_dataset = TimeSeriesDataset(X_train, y_train)
test_dataset = TimeSeriesDataset(X_test, y_test)

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False)
```

```python
n_features = train_data_processed.shape[1]
```

```python
model = BayesianLSTM(n_features=n_features, output_length=1, batch_size=BATCH_SIZE)
```

```python
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else  torch.device("mps") if torch.backends.mps.is_available() else 'cpu')
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import  mean_absolute_error
```

```python
def plot_training_history(results, ncols=2, patience=0):
    num_rows = len(results)
    fig, axes = plt.subplots(nrows=num_rows, ncols=ncols, figsize=(20, 5 * num_rows))
    if num_rows == 1:
        axes = [axes]
    if ncols == 1:
        axes = [[ax] for ax in axes]

    for ax, (key, values) in zip(axes, results.items()):
        epochs = range(1, len(values['train_loss']) + 1)
        stop_epoch = None
        if patience > 0:
            num_epochs = len(epochs)
            stop_epoch = num_epochs - patience
            print(f"Early stopping at epoch {stop_epoch}")


        train_loss = [x.cpu().numpy() if isinstance(x, torch.Tensor) else x for x in values['train_loss']]
        test_loss = [x.cpu().numpy() if isinstance(x, torch.Tensor) else x for x in values['test_loss']]


        ax[0].plot(epochs, train_loss, label='Train Loss')

        ax[0].plot(epochs, test_loss, label='Test Loss')
        ax[0].set_title(f"{key} Loss")
        ax[0].set_xlabel("Epoch")
        ax[0].set_ylabel("Loss")
        ax[0].legend()

        if stop_epoch:
            ax[0].axvline(x=stop_epoch, linestyle='--', label='Early Stopping')
            ax[0].legend()

        if ncols == 1:
            continue

        mae = [x.cpu().numpy() if isinstance(x, torch.Tensor) else x for x in values['mae']]
        ax[1].plot(epochs, mae, label='MAE')
        ax[1].set_title(f"{key} MAE")
        ax[1].set_xlabel("Epoch")
        ax[1].set_ylabel("MAE")
        ax[1].legend()

        if stop_epoch:
            ax[1].axvline(x=stop_epoch, linestyle='--', label='Early Stopping')
            ax[1].legend()

    plt.tight_layout()
    plt.show()
```

```python
def visualize_predictions(predicted_temps, actual_temps):
    dates = test_data.index

    if not isinstance(predicted_temps, np.ndarray):
        predicted_temps = np.array(predicted_temps)
    if not isinstance(actual_temps, np.ndarray):
        actual_temps = np.array(actual_temps)
    if not isinstance(dates, np.ndarray):
        dates = np.array(dates)[4:]

    plt.figure(figsize=(15, 7))
    plt.plot(dates, predicted_temps, label='Predicted Temperatures', color='red', marker='o', linestyle='-', markersize=2)
    plt.plot(dates, actual_temps, label='Actual Temperatures', marker='x', linestyle='-', markersize=2)
    plt.xlabel('Date')
    plt.ylabel('Temperature (Â°C)')
    plt.title('Comparison of Predicted and Actual Temperatures')
    plt.legend()
    plt.gcf().autofmt_xdate()
    plt.tight_layout()
    plt.show()
```

```python
def train_and_evaluate(model, train_loader, test_loader, optimizer, epochs=100, patience=10, plot=True, print_results=True):
    model.train()
    best_test_loss = float('inf')
    no_improve_epochs = 0
    early_stopping = False
    history = {
        'train_loss': [],
        'test_loss': [],
        'mae': []
    }

    for epoch in range(epochs):
        model.train()
        total_loss = 0
        for X_batch, Y_batch in train_loader:
            X_batch, Y_batch = X_batch.to(device), Y_batch.to(device)
            predictions = model(X_batch)
            loss = model.loss(predictions.view(-1), Y_batch)

            optimizer.zero_grad()

            loss.backward()
            optimizer.step()
            total_loss += loss.item()


        average_train_loss = total_loss / len(train_loader)
        history['train_loss'].append(average_train_loss)

        model.eval()
        test_loss = 0
        predictions_arr = []
        actuals = []

        with torch.no_grad():
            for X_batch, Y_batch in test_loader:
                X_batch, Y_batch = X_batch.to(device), Y_batch.to(device)
                preds = model(X_batch)
                loss = model.loss(preds.view(-1), Y_batch)
                test_loss += loss.item()

                temperature_scaler = preprocessor.named_transformers_['num']
                temp_preds = np.zeros((preds.shape[0], 9))
                temp_preds[:, 0] = preds.view(-1).cpu().numpy()
                temp_preds = temperature_scaler.inverse_transform(temp_preds)
                preds = temp_preds[:, 0]

                temp_y = np.zeros((Y_batch.shape[0], 9))
                temp_y[:, 0] = Y_batch.view(-1).cpu().numpy()
                temp_y = temperature_scaler.inverse_transform(temp_y)

                Y_batch = temp_y[:, 0]

                predictions_arr.extend(preds)
                actuals.extend(Y_batch)

        average_test_loss = test_loss / len(test_loader)

        mae = mean_absolute_error(actuals, predictions_arr)
        history['test_loss'].append(average_test_loss)
        history['mae'].append(mae)

        if print_results:
            print(f'Epoch {epoch + 1}: \n\tTrain Loss = {average_train_loss:.4f}, \n\tTest Loss  = {average_test_loss:.4f}, MAE = {mae:.4f}Â°C')

        if average_test_loss < best_test_loss:
            best_test_loss = average_test_loss
            no_improve_epochs = 0
        else:
            no_improve_epochs += 1
            if no_improve_epochs >= patience:
                if print_results:
                    print("Early stopping triggered")
                early_stopping = True
                break

    if print_results:
        print("\nTraining completed")

    if plot:
        plot_training_history({'LSTM-BN': history}, patience=patience)

    if early_stopping:
        history['train_loss'] = history['train_loss'][:-(patience)]
        history['test_loss'] = history['test_loss'][:-(patience )]
        history['mae'] = history['mae'][:-(patience)]

    if print_results:
        print(f"Test Loss: {best_test_loss:.4f}")
        print(f"Åšredni bÅ‚Ä…d bezwzglÄ™dny (MAE): {history['mae'][-1]:.2f}Â°C")

    if plot:
        visualize_predictions(predictions_arr, actuals)
    return history
```

```python
train_and_evaluate(model, train_loader, test_loader, optimizer, epochs=EPOCHS, patience=5)
```

```python
from itertools import product


def hyperparameter_search_BayesianLSTM(train_loader, test_loader):
    feature_dim = X_train.shape[2]

    hidden_dim1_options = [64, 128, 256]
    hidden_dim2_options = [16, 32, 64]

    num_layers_options = [1, 2, 3]
    lr_options = [0.0001, 0.001, 0.01]
    batch_size_options = [16, 32, 64]

    results = []

    for hidden_dim1, hidden_dim2, stacked_layers, lr, batch_size in product(hidden_dim1_options, hidden_dim2_options, num_layers_options, lr_options, batch_size_options):
        print(f'Testing configuration: hidden_dim1={hidden_dim1}, hidden_dim2={hidden_dim2}, num_layers={stacked_layers}, lr={lr}, batch_size={batch_size}')

        model = BayesianLSTM(feature_dim, 1, hidden_dim1, hidden_dim2, stacked_layers)
        optimizer = torch.optim.Adam(model.parameters(), lr=lr)

        metrics = train_and_evaluate(model, train_loader, test_loader, optimizer, epochs=50, patience=10, plot=False, print_results=False)

        results.append({
            'hidden_dim1': hidden_dim1,
            'hidden_dim2': hidden_dim2,
            'num_layers': stacked_layers,
            'lr': lr,
            'batch_size': batch_size,
            'train_loss': metrics['train_loss'][-1],
            'test_loss': metrics['test_loss'][-1],
            'mae': metrics['mae'][-1]
        })
        print(f"Test Loss: {metrics['test_loss'][-1]:.4f}, MAE: {metrics['mae'][-1]:.2f}")

    results_df = pd.DataFrame(results)
    results_df.to_csv('BayesianLSTM_hyperparameter_search_results.csv', index=False)
    print("BayesianLSTM Hyperparameter search results saved to 'BayesianLSTM_hyperparameter_search_results.csv'")
    return results_df

bayesian_lstm_results_df = hyperparameter_search_BayesianLSTM(train_loader, test_loader)
```

```python
def inverse_transform(predictions):
    temperature_scaler = preprocessor.named_transformers_['num']
    temp_preds = np.zeros((predictions.shape[0], 9))
    temp_preds[:, 0] = predictions.view(-1).detach().cpu().numpy()
    temp_preds = temperature_scaler.inverse_transform(temp_preds)
    return temp_preds[:, 0]
```

```python
N_EXPERMENTS = 100

test_uncertainty_df = pd.DataFrame()
test_uncertainty_df['date'] = test_data.index[4:]

testing_truth_df = pd.DataFrame()
testing_truth_df['date'] = test_data.index[4:]

testing_preds_df = pd.DataFrame()
testing_preds_df['date'] = test_data.index[4:]

for i in range(N_EXPERMENTS):
    predictions_all = []
    targets_all = []
    for X_batch, Y_batch in test_loader:
        X_batch, Y_batch = X_batch.to(device), Y_batch.to(device)
        predictions = inverse_transform(model(X_batch))
        predictions_all.extend(predictions)
        if i == 0:
            targets_all.extend(inverse_transform(Y_batch))
    test_uncertainty_df[f'prediction_{i}'] = predictions_all
    if i == 0:
        testing_truth_df['truth'] = targets_all
        testing_preds_df['prediction'] = predictions_all

temp_pred_df = test_uncertainty_df.filter(like='prediction', axis=1)
test_uncertainty_df['mean'] = temp_pred_df.mean(axis=1)
test_uncertainty_df['std'] = temp_pred_df.std(axis=1)

test_uncertainty_df = test_uncertainty_df[['date', 'mean', 'std']]
test_uncertainty_df
```

```python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(testing_truth_df['truth'], testing_preds_df['prediction'])
mse
```

```python
test_uncertainty_df['lower_bound'] = test_uncertainty_df['mean'] - 1.645 * test_uncertainty_df['std']
test_uncertainty_df['upper_bound'] = test_uncertainty_df['mean'] + 1.645 * test_uncertainty_df['std']
```

```python
test_uncertainty_plot_df = test_uncertainty_df.copy(deep=True)
truth_uncertainty_plot_df = testing_truth_df.copy(deep=True)
```

```python
bounds_df = pd.DataFrame()

# Using 90% confidence bounds
bounds_df['lower_bound'] = test_uncertainty_plot_df['lower_bound']
bounds_df['prediction'] = test_uncertainty_plot_df['mean']
bounds_df['real_value'] = truth_uncertainty_plot_df['truth']
bounds_df['upper_bound'] = test_uncertainty_plot_df['upper_bound']

bounds_df['contained'] = ((bounds_df['real_value'] >= bounds_df['lower_bound']) &
                          (bounds_df['real_value'] <= bounds_df['upper_bound']))

print("Proportion of points contained within 90% confidence interval:",
      bounds_df['contained'].mean())
```
